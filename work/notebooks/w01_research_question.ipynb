## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
#The Search Question: Which high-exposure content assets are showing stable or decaying search presence and should be prioritized for immediate editorial intervention?
#Unit of Analysis: A single, unique pseudonymized content item (one page).
#The Decision: How should the content operations team allocate their fixed review and optimization capacity for the upcoming production cycle?
#The Action: An editor pulls a page from the top of the prioritized queue, audits its missing information or structural issues based on the provided reason codes, and updates/refreshes the live content.
#The Cost of a Wrong Recommendation:

#False Positive (Recommending a healthy page): Wastes expensive human editorial hours and salary reviewing a page that doesn't need help, yielding zero return on investment (ROI).
#False Negative (Missing a decaying page): Allows a high-value asset to continue dropping in rankings, causing permanent traffic and revenue loss as competitors claim those search positions.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np

# Load the starter dataset
df = pd.read_csv("content_refresh_anonymized.csv")

# Total Content Inventory Size
total_pages = len(df)

# Scope of the Decay Problem (Handling the Label Trap safely)
declining_pages = len(df[df['trend_direction'] == 'down'])
pct_declining = (declining_pages / total_pages) * 100

# High-Exposure Baseline (Filtering out unexposed pages & avg_position = 0 gotcha)
valid_exposure = df[(df['impressions_90d'] >= 500) & (df['avg_position'] > 0)]
high_exposure_count = len(valid_exposure)
pct_high_exposure = (high_exposure_count / total_pages) * 100

print("--- DATA INVESTIGATION VERDICT ---")
print(f"Total Content Inventory (Grain: Unique Page): {total_pages:,}")
print(f"Active Decay Base Rate (trend_direction == 'down'): {declining_pages:,} ({pct_declining:.1f}%)")
print(f"High-Visibility Target Inventory (Impressions >= 500 & Position > 0): {high_exposure_count:,} ({pct_high_exposure:.1f}%)")
```
